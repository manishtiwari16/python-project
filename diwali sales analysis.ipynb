```python
#this shows what product has sold maximum times
ax=sns.barplot(x=list(df['Product_ID'].value_counts().head().keys()),y=list(df['Product_ID'].value_counts().head()),data=df)
for bar in ax.containers:
    ax.bar_label(bar)
plt.show()
            
```

```python
top_cust=df.groupby(['Cust_name','Age Group'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head()
top_cust
```

```python
#these are our top 5 customers who ordered the most
ax=sns.barplot(x='Cust_name',y='Orders',hue='Age Group',data=top_cust)
for bar in ax.containers:
    ax.bar_label(bar)
plt.show()
```
